```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
#Load the data
aviation = pd.read_csv('project3_data.csv', encoding='latin1', low_memory=False)
```

```python
cols = ("TotalFatalInjuries",'WeatherCondition','BroadPhaseofFlight','PurposeofFlight', "Lethality", "NumberofEngines","EngineType")
df = pd.DataFrame(aviation, columns=cols);
df.shape
```

```python
df = df.dropna()
df.shape
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
df.shape, train_df.shape, test_df.shape
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

train_df["TotalFatalInjuries_code"] = le.fit_transform(train_df["TotalFatalInjuries"])
train_df[["TotalFatalInjuries_code","TotalFatalInjuries"] ]
```

```python
train_df["Lethality_code"] = le.fit_transform(train_df["Lethality"])
train_df[["Lethality_code","Lethality"] ]
```

```python
train_df["WeatherCondition_code"] = le.fit_transform(train_df["WeatherCondition"])
train_df[["WeatherCondition_code","WeatherCondition"] ]
```

```python
train_df["BroadPhaseofFlight_code"] = le.fit_transform(train_df["BroadPhaseofFlight"])
train_df[["BroadPhaseofFlight_code","BroadPhaseofFlight"] ]
```

```python
train_df["PurposeofFlight_code"] = le.fit_transform(train_df["PurposeofFlight"])
train_df[["PurposeofFlight_code","PurposeofFlight"] ]
```

```python
train_df["NumberofEngines_code"] = le.fit_transform(train_df["NumberofEngines"])
train_df[["NumberofEngines_code","NumberofEngines"] ]
```

```python
train_df["EngineType_code"] = le.fit_transform(train_df["EngineType"])
train_df[["EngineType_code","EngineType"] ]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(train_df[["TotalFatalInjuries_code","BroadPhaseofFlight_code","WeatherCondition_code", "Lethality_code", "NumberofEngines_code","EngineType_code", "PurposeofFlight_code"]])
scaled_train_df = scaler.transform(train_df[["TotalFatalInjuries_code","BroadPhaseofFlight_code","WeatherCondition_code", "Lethality_code", "NumberofEngines_code","EngineType_code", "PurposeofFlight_code"]])
scaled_test_df = scaler.transform(train_df[["TotalFatalInjuries_code","BroadPhaseofFlight_code","WeatherCondition_code", "Lethality_code", "NumberofEngines_code","EngineType_code", "PurposeofFlight_code"]])

scaled_train_df, scaled_test_df
#test_df[["TotalFatalInjuries_code","WeatherCondition_code"]]
```

```python
# Now use Standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(train_df[["TotalFatalInjuries_code","BroadPhaseofFlight_code","WeatherCondition_code", "Lethality_code", "NumberofEngines_code","EngineType_code", "PurposeofFlight_code"]])
scaled_train_df = scaler.transform(train_df[["TotalFatalInjuries_code","BroadPhaseofFlight_code","WeatherCondition_code", "Lethality_code", "NumberofEngines_code","EngineType_code", "PurposeofFlight_code"]])
scaled_test_df = scaler.transform(train_df[["TotalFatalInjuries_code","BroadPhaseofFlight_code","WeatherCondition_code", "Lethality_code", "NumberofEngines_code","EngineType_code", "PurposeofFlight_code"]])

scaled_train_df, scaled_test_df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(scaled_train_df)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])
principalDf
```

```python
train_df[["Lethality"]]
```

```python
finalDf = pd.concat([principalDf, train_df[["Lethality"]]], axis = 1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('pc1', fontsize = 15)
ax.set_ylabel('pc2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0.0,1.0]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Lethality'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'pc1']
               , finalDf.loc[indicesToKeep, 'pc2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
pca.explained_variance_ratio_
```

```python
#It takes 6 components to cover 97% variance
pca = PCA(n_components=6)
principalComponents = pca.fit_transform(scaled_train_df)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6'])
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(scaled_train_df)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2', 'pc3'])
principalDf
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.cluster import KMeans
principalDf.shape
```

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit_predict(principalComponents)
    wcss.append(kmeans.inertia_)
```

```python
wcss    
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,10), wcss, marker='o', linestyle = '--') 
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sum of Squares (WCSS)')
plt.title('Kmeans Clustering w/ Prncipal Components')
plt.show()
```
